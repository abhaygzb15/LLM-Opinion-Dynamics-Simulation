```python
import os
import pandas as pd
import plotly.express as px
from IPython.display import display, HTML

# 2. Define Input and Output Folders
INPUT_FOLDER = "simulation_csvs"
OUTPUT_FOLDER = "simulation_results"

# Create the output folder if it doesn't exist
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# 3. Check if input folder exists and has files
if not os.path.exists(INPUT_FOLDER) or not os.listdir(INPUT_FOLDER):
    print(f"Please create a folder named '{INPUT_FOLDER}' and upload your CSV files into it.")
else:
    # 4. Loop through every CSV file in the input folder
    for filename in os.listdir(INPUT_FOLDER):
        if filename.endswith(".csv"):
            filepath = os.path.join(INPUT_FOLDER, filename)
            experiment_name = filename.replace('.csv', '')
            
            # Create a dedicated subfolder for this specific experiment's outputs
            subfolder_path = os.path.join(OUTPUT_FOLDER, experiment_name)
            os.makedirs(subfolder_path, exist_ok=True)
            
            print(f"\n{'='*60}")
            print(f"Processing Experiment: {experiment_name}")
            print(f"{'='*60}")
            
            # Load Data
            df = pd.read_csv(filepath)
            
            # Validate Columns
            required_columns = ['Round', 'Agent_ID', 'Stance_Score', 'Persona']
            if not all(col in df.columns for col in required_columns):
                print(f"Error: {filename} is missing required columns. Skipping.")
                continue

            # --- A. CALCULATE STATISTICS ---
            system_metrics = df.groupby('Round')['Stance_Score'].agg(['mean', 'var', 'std']).reset_index()
            system_metrics = system_metrics.rename(columns={
                'mean': 'System_Average', 
                'var': 'System_Variance', 
                'std': 'Standard_Deviation'
            })
            
            # Save Stats to CSV in subfolder
            stats_csv_path = os.path.join(subfolder_path, f"{experiment_name}_statistics.csv")
            system_metrics.to_csv(stats_csv_path, index=False)
            
            print(f"\n Mathematical Proof of System Dynamics (Saved to {experiment_name}_statistics.csv)")
            display(system_metrics.round(2)) # Pretty print table in Colab

            # --- B. GRAPH 1: OPINION DRIFT ---
            fig_drift = px.line(
                df, x="Round", y="Stance_Score", color="Agent_ID", hover_name="Persona", markers=True,
                title=f"Opinion Drift Over Time ({experiment_name})",
                labels={"Round": "Debate Round", "Stance_Score": "Stance Score (1=Against, 10=For)", "Agent_ID": "Agent"}
            )
            fig_drift.update_layout(
                template="plotly_dark", yaxis=dict(range=[0, 10.5], dtick=2), 
                xaxis=dict(dtick=1), height=500, hovermode="x unified"
            )
            
            # Save Graph 1
            drift_img_path = os.path.join(subfolder_path, f"{experiment_name}_drift.png")
            fig_drift.write_image(drift_img_path)
            
            # Display Graph 1 Inline
            fig_drift.show()

            # --- C. GRAPH 2: SYSTEM VOLATILITY ---
            fig_variance = px.line(
                system_metrics, x="Round", y="Standard_Deviation", markers=True,
                title=f"System Volatility and Polarization ({experiment_name})",
                labels={"Standard_Deviation": "Standard Deviation (Higher = Polarized)", "Round": "Debate Round"}
            )
            fig_variance.update_layout(
                template="plotly_dark", yaxis=dict(range=[0, 5], dtick=1), 
                xaxis=dict(dtick=1), height=400
            )
            # Update line color to stand out
            fig_variance.update_traces(line_color='#FF4B4B') 
            
            # Save Graph 2
            var_img_path = os.path.join(subfolder_path, f"{experiment_name}_volatility.png")
            fig_variance.write_image(var_img_path)
            
            # Display Graph 2 Inline
            fig_variance.show()
            
            print(f"Successfully saved all assets for '{experiment_name}' to: {subfolder_path}\n")

print("\n ALL FILES PROCESSED SUCCESSFULLY!")
```
